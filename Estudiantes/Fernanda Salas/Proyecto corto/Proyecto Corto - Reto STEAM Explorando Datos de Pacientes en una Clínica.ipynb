## Cargar y Leer los Datos de Pacientes

````python
import json

with open("clinica.json", encoding="utf-8") as archivo:
    datos_pacientes = json.load(archivo)
    
with open("clinica_intermedio.json", encoding = "utf'8") as archivo_intermedio:
    datos_inter_pacientes = json.load(archivo_intermedio)
    
registro_clinica = datos_pacientes + datos_inter_pacientes

print(f'registro de pacientes: ', registro_clinica)

````

## Desafío 1: Análisis Científico - Reporte de Enfermedades Tratadas

Tu primer reto es analizar los datos de los pacientes para generar un reporte de las enfermedades tratadas en la clínica. Utiliza listas y ciclos para contar cuántos pacientes han sido tratados por cada enfermedad.

### Instrucciones:

1. Crea una lista vacía donde almacenarás las enfermedades.
2. Recorre los datos de los pacientes y extrae la enfermedad de cada paciente.
3. Cuenta cuántos pacientes han sido tratados por cada enfermedad.
4. Visualiza los resultados de manera clara utilizando la técnica que elijas (gráficos o tablas).

### Preguntas:

### **Paso a Paso para el Estudiante**

#### **Paso 1: Crear las Listas**

### 

1. **Objetivo:** Necesitamos dos listas para almacenar:
    
    - Los nombres de las enfermedades.
    - La cantidad de veces que cada enfermedad aparece.
2. **Instrucción:**
    
    - Crea una lista vacía llamada `lista_enfermedades`.
    - Crea otra lista vacía llamada `lista_cantidades`.

````python
from collections import Counter

# Lista para almacenar las enfermedades y sus cantidades

lista_enfermedades = []

for paciente in registro_clinica:
    lista_enfermedades.append(paciente[4]) #posicion 4 

conteo_enfermedad = Counter(lista_enfermedades) 

#acomodando para visualizar mejor
#enfermedad en posicion 0, cantidad en posicion 1 de la tupla

print("Cantidad de pacientes con enfermedad: ")

for enfermedad, cantidad in conteo_enfermedad.most_common():
    print(f"{enfermedad}: {cantidad} pacientes")
    
````

````python
#Unificar diagnósticos para mejorar el conteo

estandarizacion = {
    "dolor de espalda": "dolor",
    "presión alta" : "presion",
    "presion" : "presion",
}

conteo_corregido = Counter()

for enfermedad, cantidad in conteo_enfermedad.items():
    estandar_enfermedades = estandarizacion.get(enfermedad, enfermedad)
    conteo_corregido[estandar_enfermedades] += cantidad

print("Enfermedades diagnosticadas estandarizadas: ")

for enfermedad, cantidad in conteo_corregido.most_common():
    print(f"{enfermedad}: {cantidad} pacientes")
````

**¿Qué enfermedades son las más tratadas en la clínica?**

*Las enfermedades más tratadas son gripe y dolor*

**¿Hay alguna enfermedad que te sorprenda por su prevalencia?**

*Muchos síntomas son tratadas con fármacos de venta libre antes de ir al médico, por lo cual una gripe o dolor puede significar alguna dolencia mayor que requiere de atención especializada o necesidad de una incapacidad justificada; en otros casos como la ansiedad por ejemplo, se diagnosticó solo dos veces aunque probablemente muchas personas la padezcan; pero en este tipo de padecimientos se suele normalizar los síntomas y no ir al médico.*

#### **Paso 2: Recorrer los Datos de los Pacientes**

### 

1. **Objetivo:** Vamos a analizar cada paciente para extraer su enfermedad.
    
2. **Instrucción:**
    
    - Usa un bucle `for` para recorrer cada `paciente` en `datos_pacientes` (ya está definida en el código como una lista de listas).
    - Dentro del bucle, asigna la enfermedad del paciente (índice `4` de cada lista) a una variable llamada `enfermedad`.

#### **Paso 3: Verificar si la Enfermedad ya Está en la Lista**

1. **Objetivo:** Verificar si la enfermedad ya fue registrada.
    
2. **Instrucción:**
    
    - Usa un `if` para verificar si la `enfermedad` ya está en `lista_enfermedades`.
    - Si **no** está en `lista_enfermedades`, agrega la enfermedad a `lista_enfermedades` y añade `1` a `lista_cantidades` para contarla por primera vez.
    - Si **sí** está en `lista_enfermedades`, encuentra la posición (`indice`) de la enfermedad en `lista_enfermedades` y aumenta en `1` el valor en `lista_cantidades` en la misma posición.

````python
# Listas para almacenar enfermedades y sus cantidades
lista_enfermedades_estandarizada = list(set(estandarizacion.get(e, e) for e in lista_enfermedades)) #uso de set para eliminar duplicas y de get para buscar en estandarizacion si existe la enfermedad y la devuelve estandarizada.
lista_cantidades = [0] * len(lista_enfermedades_estandarizada)

# Recorrer los datos de los pacientes
for paciente in registro_clinica:
    enfermedad = paciente[4]
    enfermedad_final = estandarizacion.get(enfermedad, enfermedad)
    # Pista: Si la enfermedad ya está, usa `index` para encontrar su posición en lista_enfermedades y suma 1 en lista_cantidades en esa posición.
    if enfermedad_final in lista_enfermedades_estandarizada:
        indice = lista_enfermedades_estandarizada.index(enfermedad_final)
        lista_cantidades[indice] += 1
    # Pista: Si la enfermedad no está, agrégala a lista_enfermedades y empieza el conteo en lista_cantidades.
    elif enfermedad_final not in lista_enfermedades_estandarizada:
        lista_enfermedades_estandarizada.append(enfermedad_final)
        lista_cantidades.append(1)
        
print("\nEnfermedades tratadas: ")
for i in range(len(lista_enfermedades_estandarizada)):
    print(f"{lista_enfermedades_estandarizada[i]}: {lista_cantidades[i]} pacientes")
````

#### **Paso 4: Imprimir el Reporte Final**

1. **Objetivo:** Mostrar el número de pacientes con cada enfermedad.
    
2. **Instrucción:**
    
    - Usa un bucle `for` para recorrer la lista `lista_enfermedades`.
    - En cada iteración, imprime la enfermedad y el número de pacientes (que está en `lista_cantidades` en la misma posición).

````python
print("\nReporte de enfermedades tratadas: ")
for i in range(len(lista_enfermedades_estandarizada)):
    print(f"{lista_enfermedades_estandarizada[i]}: {lista_cantidades[i]} pacientes")
````

## Desafío 2: Exploración Tecnológica - Reporte de Medicamentos Recetados

Ahora que ya tienes el reporte de las enfermedades tratadas, es momento de explorar qué medicamentos se están recetando en la clínica. Tu tarea es generar un reporte similar al anterior, pero esta vez debes contar cuántos pacientes han recibido cada medicamento.

### Instrucciones:

1. Almacena los medicamentos recetados en una lista.
2. Recorre los datos y cuenta cuántos pacientes han recibido cada medicamento.
3. Visualiza los resultados de manera creativa.

### Preguntas:

- ¿Qué medicamentos son los más recetados?
- ¿Cómo podrías optimizar el uso de medicamentos en la clínica?

Resultados esperados:


````python
# Listas para almacenar los medicamentos y sus frecuencias
medicamentos = []
frecuencias = []

for paciente in registro_clinica:
    medicamentos.append(paciente[5])
    
conteo_medicamentos = Counter(medicamentos) 

print("Cantidad de pacientes por medicamento: ")

for medicamento, cantidad in conteo_medicamentos.most_common():
    print(f"{medicamento}: {cantidad} pacientes")

````

````python
# Unificar nombres de medicamentos
estandarizacion_meds = {
    "acetaminofen": "paracetamol",
    "antiácido": "antiacido",
}

medicamentos_estandarizados = list(set(estandarizacion_meds.get(m, m) for m in medicamentos))

frecuencias = [0] * len(medicamentos_estandarizados)

for paciente in registro_clinica:
    medicamento = paciente[5]  # medicamento
    medicamento_final = estandarizacion_meds.get(medicamento, medicamento)

    if medicamento_final in medicamentos_estandarizados:
        indice = medicamentos_estandarizados.index(medicamento_final)
        frecuencias[indice] += 1

    elif medicamento_final not in medicamentos_estandarizados:
        medicamentos_estandarizados.append(medicamento_final)
        frecuencias.append(1)

print("\nReporte de medicamentos recetados:")
for i in range(len(medicamentos_estandarizados)):
    print(f"{medicamentos_estandarizados[i]}: {frecuencias[i]} pacientes")
````

````python
#Mostrando los datos en una tabla:
#pip install pandas en terminal

import pandas as pd
from tabulate import tabulate

#
df_medicamentos = pd.DataFrame({"Medicamento": medicamentos_estandarizados, "Pacientes": frecuencias})

# Mostrar tabla en formato bonito
print(tabulate(df_medicamentos, headers="keys", tablefmt="fancy_grid"))
````

````python
#Mostrando los datos con nubes: 
#pip install wordcloud en terminal

from wordcloud import WordCloud
import matplotlib.pyplot as plt

#Crear un diccionario {medicamento: frecuencia}
frecuencia_medicamentos = dict(zip(medicamentos_estandarizados, frecuencias))

#Generar la nube de palabras
wordcloud = WordCloud(width=800, height=400, background_color="black", colormap="Dark2")
wordcloud.generate_from_frequencies(frecuencia_medicamentos)

#Mostrar la nube de palabras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # Ocultar ejes
plt.title("Nube de Palabras - Medicamentos Recetados")
plt.show()
````
